```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.cars-data.com/en/all-cars/page1.html' # Initial URL for scraping

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
urls =['https://www.cars-data.com/en/all-cars/page1.html']
for i in range(1, 98):
    url = 'https://www.cars-data.com/en/all-cars/page' + str(i) + '.html'

    page = requests.get(url)
    print(page)

    soup = BeautifulSoup(page.text, 'html')


    nextpage_link = soup.find('a', text = 'next')

    if nextpage_link:
        nextpage_url = nextpage_link.get('href')
        urls.append(nextpage_url)
        print(nextpage_url)
```

```python
print(urls)
```

```python
names = []
year_range = []
bodystyles = []
no_of_doors = []
n = 1

for link in urls:
    page = requests.get(link)
    soup = BeautifulSoup(page.text, 'html')
    box = soup.find_all('div', class_ = 'col-4')
    
    for each in box:
  # Check if there are at least 2 lines of text before splitting
      if len(each.text.split('\n')) >= 2:
        name = each.text.split('\n')[0]
        year = each.text.split('\n')[1][:11]
        try:
          # Attempt to split using ', ' (handle potential exceptions)
          bodystyle = each.text.split(', ')[1]
        except IndexError:
          # If there's no comma, assign an empty string or handle differently
          bodystyle = ""
        doors = each.text.split('\n')[1].split(', ')[0][11:]
      else:
        # Handle cases with less than 2 lines
        name = ""
        year = ""
        bodystyle = ""
        doors = ""

      names.append(name)
      year_range.append(year)
      bodystyles.append(bodystyle)
      no_of_doors.append(doors)
      print(f"Done {link} successfully")
```

```python
print(names)
```

```python
data = {
    "Car Names":names,
    "Years of Production": year_range,
    "Number Of Doors":no_of_doors,
    "BodyStyles":bodystyles
}
```

```python
df = pd.DataFrame(data)
```
